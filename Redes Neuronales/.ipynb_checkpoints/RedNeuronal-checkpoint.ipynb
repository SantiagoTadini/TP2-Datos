```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
train = pd.read_csv("../Features/data_train.csv")
test = pd.read_csv("../Features/data_test.csv")
```

```python
# x son los features, y es el target
train = train.drop([], axis=1)

x, y = train.iloc[:, :-1].values, train.iloc[:, -1].values
```

```python
SCALING
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.20)
```

```python
x.shape
```

```python
def create_model(optimizer='adam', dropout=0.1, init='uniform', n_nodes=256):
    model = Sequential()
    model.add(Dense(n_nodes, input_shape=(x.shape[1],), activation='relu', kernel_initializer=init))
    model.add(Dropout(dropout), )
    model.add(Dense(n_nodes, activation='relu'))
    model.add(Dropout(dropout), )
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=create_model)
```

```python
param_grid = {
    'epochs': [100, 1000],
    'batch_size': [1000, 5000, 10000],
    'n_nodes1': [32, 256, 512],
    'n_nodes2': [32, 256, 512],
    'optimizer': ['Adamax'],
    'dropout': [0.5, 0.4, 0.3, 0.2, 0.1, 0],
    'init': ['normal']
}
```

```python
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=4)
grid_result= grid.fit(X_train, Y_train)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

```python
test_accuracy = grid.score(X_test, Y_test)
```

```python
accuracy = red_neuronal.evaluate(X_test, Y_test)
```
